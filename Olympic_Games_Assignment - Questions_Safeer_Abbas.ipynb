```python
## Importing Required Libraries
# !pip install pandas
import pandas as pd

## Reading the dataset
df = pd.read_csv("athlete_events.csv")
```

### DATA CLEANING GOES BELOW IT

```python
df.head()
```

```python
df.info()
```

```python
# Remove duplicate rows from the DataFrame
df = df.drop_duplicates()
```

```python
# Check for duplicate rows in the DataFrame
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
```

```python
# Rename the 'Sex' column to 'Gender'
df = df.rename(columns={'Sex': 'Gender'})
```

```python
df.head(5)
```

```python
# Change the members of Gender from M -> Male, F -> Female
df['Gender'] = df['Gender'].replace({'M': 'Male', 'F': 'Female'})
df.head(10)
```

```python
# Impute NaN / Null values in the 'Medal' column with "No Medal"
df['Medal'] = df['Medal'].fillna("No Medal")
# Display the entire 'Medal' column
print(df['Medal'])
```

#### OBJECTIVE QUESTIONS

##### Q1: Which gender has won most medals across all Medal Categories?

```python
## Python Pandas Solution

# Group by 'Gender' and count the number of medals
medal_counts = df['Gender'].value_counts()

# Display the counts
print(medal_counts)
```

```python
## Relevant Visualization Using Python
# !pip install matplotlib
# !pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt
medal_counts.plot(kind='bar', color=['blue', 'pink'])
plt.title('Total Medals by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Medals')
plt.xticks(rotation=0)
plt.show()
```

##### Q2: Which country has won the most gold medals? Put Top 5

```python
## Python Pandas Solution

# Filter the DataFrame for gold medals
gold_medals = df[df['Medal'] == 'Gold']

# Group by 'Team' and count the number of gold medals
gold_medal_counts = gold_medals['Team'].value_counts().head(5)

# Display the counts
print(gold_medal_counts)
```

```python
## Relevant Visualization Using Python
# Create a bar plot for the top 5 countries with the most gold medals
gold_medal_counts.plot(kind='bar', color='gold')
plt.title('Top 5 Countries with Most Gold Medals')
plt.xlabel('Country')
plt.ylabel('Number of Gold Medals')
plt.xticks(rotation=45)
plt.show()
```

##### Q3: What is the trend of Number of Participants in Each Olympic Years?

```python
## Python Pandas Solution
# Group by 'Year' and count the number of participants
participants_trend = df.groupby('Year')['ID'].nunique()

# Display the counts
print(participants_trend)
```

```python
## Relevant Visualization Using Python
# Create a line plot for the trend of participants over the years
participants_trend.plot(kind='line', marker='o', color='blue')
plt.title('Trend of Number of Participants in Each Olympic Year')
plt.xlabel('Year')
plt.ylabel('Number of Participants')
plt.xticks(participants_trend.index, rotation=90)
plt.grid()
plt.show()
```

```python
# Group by 'Year' and 'Team' and count the number of participants
participants_trend = df.groupby(['Year', 'Team'])['ID'].nunique().unstack(fill_value=0)

# Select the top 20 countries based on total participants
top_20_countries = participants_trend.sum().nlargest(20).index
participants_trend_top_20 = participants_trend[top_20_countries]

# Display the counts for top 20 countries
print(participants_trend_top_20)

## Relevant Visualization Using Python
import matplotlib.pyplot as plt

# Create a line plot for the trend of participants over the years for the top 20 countries
participants_trend_top_20.plot(kind='line', marker='o')
plt.title('Trend of Number of Participants in Each Olympic Year by Top 20 Countries')
plt.xlabel('Year')
plt.ylabel('Number of Participants')
plt.xticks(rotation=90)
plt.grid()
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust legend position
plt.tight_layout()  # Adjust layout to make room for the legend
plt.show()
```

##### Q4: Which Sport had Highest share of Silver Medals in 2014 Olympics?

```python
# Filter the DataFrame for the year 2014 and silver medals
silver_medals_2014 = df[(df['Year'] == 2014) & (df['Medal'] == 'Silver')]

# Group by 'Sport' and count the number of silver medals
silver_medal_counts = silver_medals_2014['Sport'].value_counts()

# Identify the sport with the highest number of silver medals
highest_silver_sport = silver_medal_counts.idxmax() #this idxmax confused alot
highest_silver_count = silver_medal_counts.max()

# Display the result
print(f"The sport with the highest share of silver medals in the 2014 Olympics is '{highest_silver_sport}' with {highest_silver_count} silver medals.")
```

```python
## Relevant Visualization Using Python
# Create a bar plot for the number of silver medals by sport
silver_medal_counts.plot(kind='bar', color='silver')
plt.title('Number of Silver Medals by Sport in 2014 Olympics')
plt.xlabel('Sport')
plt.ylabel('Number of Silver Medals')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

##### Q5: What is the distribution of age for each medal type excluding "No Medal"

### Question 2: Which countries have shown the most improvement in their medal counts over the years, and what might be the reasons behind this trend?
Investigate the countries that have increased their medal counts significantly over the decades and discuss potential factors contributing to their success.

```python
## Python Pandas Solution
# Group by 'Year' and 'Team' and count the number of medals
medal_counts = df.groupby(['Year', 'Team'])['Medal'].count().unstack(fill_value=0)

# Calculate the change in medal counts for each country over the years
medal_change = medal_counts.diff().fillna(0)

# Identify the countries with the most improvement in the latest Olympic year (2016)
latest_year = medal_change.loc[2016]
most_improved_countries = latest_year[latest_year > 0].nlargest(10)

# Display the most improved countries
print("Countries with the most improvement in medal counts in 2016:")
print(most_improved_countries)
```

```python
## Relevant Visualization Using Python
# Create a line plot for the medal counts over the years for the most improved countries
plt.figure(figsize=(12, 6))
for country in most_improved_countries.index:
    plt.plot(medal_counts.index, medal_counts[country], marker='o', label=country)

plt.title('Medal Counts Over the Years for Most Improved Countries')
plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.xticks(medal_counts.index, rotation=90)
plt.legend(title='Country')
plt.grid()
plt.tight_layout()
plt.show()
```

### Question 3: How does the distribution of medals vary between summer and winter Olympic games?
Examine the differences in medal distribution across various sports in the summer and winter Olympics, and discuss any patterns or anomalies you observe.

##### Question : Which Country has the tallest athletes in Latest Olympics? Display Top 5 (Country Names)

```python
## Python Pandas Solution
# Load the main dataset
df = pd.read_csv("athlete_events.csv")

# Load the NOC Regions dataset
noc_regions = pd.read_csv("NOC_Regions.csv")
```

```python
# Display the first few rows of the NOC Regions dataset to understand its structure
print(noc_regions.head())
```

```python
# Merge the datasets on the NOC code (assuming 'NOC' is the common column)
merged_df = pd.merge(df, noc_regions, left_on='NOC', right_on='NOC', how='left')

# Display the first few rows of the merged DataFrame to verify the merge
print(merged_df.head())
```

```python
# Find the tallest athletes in the latest Olympics (2016)
tallest_athletes = merged_df[merged_df['Year'] == 2016].nlargest(5, 'Height')
print(tallest_athletes)
```

```python
# Display the top 5 countries with the tallest athletes
print("Top 5 Countries with the Tallest Athletes in the Latest Olympics (2016):")
print(tallest_athletes[['Team', 'Height']])
```
